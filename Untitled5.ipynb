```python
import pandas as pd

df = pd.read_excel("retail_sales_forcasting.xlsx")
df['Date'] = pd.to_datetime(df['Date'])
```

```python
from prophet import Prophet

prophet_df = df[['Date', 'Sales (₹)']]

prophet_df = prophet_df.rename(columns={
    'Date': 'ds',
    'Sales (₹)': 'y'
})
```

```python
model = Prophet()
model.fit(prophet_df)
```

```python
future = model.make_future_dataframe(periods=30)

forecast = model.predict(future)
```

```python
fig = model.plot(forecast)
```

```python
from statsmodels.tsa.seasonal import seasonal_decompose
import matplotlib.pyplot as plt

# Ensure your 'Date' is the index for the decomposition to work
df.set_index('Date', inplace=True)

# We use 'additive' because retail sales often grow linearly
# 'period' depends on your data: 7 for daily data (weekly cycle), 12 for monthly
decomposition = seasonal_decompose(df['Sales (₹)'], model='additive', period=7)

# Plot the components
fig = decomposition.plot()
plt.show()
```

```python
from statsmodels.tsa.stattools import adfuller

# Run the test
result = adfuller(df['Sales (₹)'])

print(f'ADF Statistic: {result[0]}')
print(f'p-value: {result[1]}')

# This is the part that was likely cut off in your image
if result[1] <= 0.05:
    print("Conclusion: Data is stationary (Reject H0)")
else:
    print("Conclusion: Data is non-stationary (Fail to reject H0). Transformation needed.")
```

```python
# Create a stationary version of your data
df['Sales_Diff'] = df['Sales (₹)'].diff().dropna()

# Plot it to see if the trend is gone
df['Sales_Diff'].plot(title="Stationary Data (Differenced)")
```

```python
# Create future dates (e.g., next 30 days)
future = model.make_future_dataframe(periods=30)

# Generate the forecast
forecast = model.predict(future)

# Look at the predicted values
print(forecast[['ds', 'yhat', 'yhat_lower', 'yhat_upper']].tail())
```
